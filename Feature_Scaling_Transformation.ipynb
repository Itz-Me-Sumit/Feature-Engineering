```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler , StandardScaler , RobustScaler
```

```python
df = pd.read_csv("housing.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df=df[['area','bedrooms','bathrooms','stories','price']]
df.head()
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data , columns = df.columns)
scaled_df.head()
```

```python
plt.figure(figsize=(12, 5))

# 1st subplot
plt.subplot(1, 2, 1)
sns.kdeplot(df["area"], fill=True, alpha=0.4, color="blue", linewidth=2)
plt.title("Original Area")

# 2nd subplot
plt.subplot(1, 2, 2)
sns.kdeplot(scaled_df["area"], fill=True, alpha=0.4, color="green", linewidth=2)
plt.title("MinMax Scaled Area")

plt.tight_layout()
plt.show()
```

```python
std_scaler = StandardScaler()
std_scaled = std_scaler.fit_transform(df)
std_scaled_df = pd.DataFrame(std_scaled , columns=df.columns)
std_scaled_df.head()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.kdeplot(df["area"] , fill=True , color = "blue" , alpha=0.4 , linewidth=2)
plt.title("Original area KDE plot")

plt.subplot(1,2,2)
sns.kdeplot(std_scaled_df["area"] , fill=True , color = "green" , alpha=0.4 , linewidth=2)
plt.title("Standard Scaled area KDE plot")

plt.show()
```

```python
robust_scaler = RobustScaler()
robust_scaled = robust_scaler.fit_transform(df)
robust_scaled_df = pd.DataFrame(robust_scaled , columns = df.columns)
robust_scaled_df.head()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.kdeplot(df["area"] , fill=True , color="blue" , alpha=0.4 , linewidth=2)
plt.title("Original area KDE plot")

plt.subplot(1,2,2)
sns.kdeplot(df["area"] , fill=True , color="red" , alpha=0.4 , linewidth=2)
plt.title("Robust Scaled area KDE plot")
```
